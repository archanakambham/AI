```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python

### Vocabulary size-dictionary size
voc_size=10000
```

```python
#step1- # one hot representation of all the words

onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
# 1670- the, 8136-glass, 5324-0f, 453-milk gives the index representation of all the words in a sentence
```

```python
# step2- pass this one hot to an embedding layer for vector representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# providing the no. of features- each word will be represented by 10 vectors in an array
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python

print(model.predict(embedded_docs)) #converted to emdedding vectors
```
